### State Wise Sum of Salary, Hra, Conv, Total

```python
df.groupby('state')[['salary','hra','conv','total']].sum()
```

### Bar Plot with State as X-axis and All Sum Values as Bars

```python
sum_state = df.groupby('state')[['salary','hra','conv','total']].sum()
x = np.arange(8)
plt.bar(x-0.3, sum_state['salary'], width=0.2, label='salary')
plt.bar(x-0.1, sum_state['hra'], width=0.2, label='hra')
plt.bar(x+0.1, sum_state['conv'], width=0.2, label='conv')
plt.bar(x+0.3, sum_state['total'], width=0.2, label='total')
plt.title('State Wise Sum of Salary, Hra, Conv & Total',fontsize=15)
plt.xlabel('States',fontsize=14)
plt.ylabel('Sum Values',fontsize=14)
plt.xticks(range(8), mean_state.index, rotation=90)
plt.legend()
plt.show()
```
