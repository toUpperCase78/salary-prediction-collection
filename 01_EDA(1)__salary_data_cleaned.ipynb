```python
# -1 should mean there is no competitor
print(salary_data['Competitors'].nunique(), "unique competitor groups")
print("\n#### COUNTS OF COMPETITORS ####")
print(salary_data['Competitors'].value_counts()[:20])
```

```python
competitors = {}
for comp in salary_data['Competitors']:
    if comp == "-1":
        continue
    comp_split = comp.split(', ')
    for one_comp in comp_split:
        if one_comp in competitors.keys():
            competitors[one_comp] += 1
        else:
            competitors[one_comp] = 1
sr_comp = pd.Series(competitors).sort_values(ascending=False)
print(len(sr_comp), "unique competitors")
print("\n#### COUNTS OF COMPETITORS ####")
print(sr_comp)
```
